# Checking values in IDs for consistency across datasets

```python
 
customer_ids = set(customer_df['customer_id'])
product_ids = set(product_df['product_id'])
device_ids = set(device_df['device_id'])
date_ids = set(date_df['date_id'])
 
inconsit = [] # Getting unique IDs, then making a list to store inconsistencies
 
customerID_missing = set(facts_df['customer_id']) - customer_ids
if customerID_missing:
    inconsit.append(f'Missing customer_ids in customer_table: {sorted(customerID_missing)}')

productID_missing = set(facts_df['product_id']) - product_ids
if productID_missing:
    inconsit.append(f'Missing product_ids in product_table: {sorted(productID_missing)}')

deviceID_missing = set(facts_df['device_id']) - device_ids
if deviceID_missing:
    inconsit.append(f'Missing device_ids in device_table: {sorted(deviceID_missing)}')

dateID_missing = set(facts_df['date_id']) - date_ids
if dateID_missing:
    inconsit.append(f'Missing date_ids in date_table: {sorted(dateID_missing)}')

if inconsit:
    print('Inconsistencies found:')
    for i in inconsit:
         print(i)
else:
     print('All ID columns are consistent across datasets.')
```

# Generating abandonment flag

```python
facts_df['is_abandoned'] = facts_df['abandon_date'].notna().astype(int)
```

```python
abandon_stats = facts_df['is_abandoned'].value_counts()
abandon_percent = facts_df['is_abandoned'].value_counts(normalize=True) * 100

print("Abandonment Statistics:")
print(f"Completed Carts (0): {abandon_stats[0]} ({abandon_percent[0]:.2f}%)")
print(f"Abandoned Carts (1): {abandon_stats[1]} ({abandon_percent[1]:.2f}%)")
print(f"\nTotal Sessions: {len(facts_df)}")
print(f"Abandonment Rate: {abandon_percent[1]:.2f}%")
```

# Using OS module to save cleaned dataframes to CSV files and exporting

```python
import os

output_dir = "C:\\Users\\ADMIN\\Desktop\\Cart Abandonment\\Cleaned_Datasets"

# Create the directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created directory: {output_dir}")

customer_path = os.path.join(output_dir, "customer_table.csv")
date_path = os.path.join(output_dir, "date_table.csv")
device_path = os.path.join(output_dir, "device_table.csv")
facts_path = os.path.join(output_dir, "fact_table.csv")
product_path = os.path.join(output_dir, "product_table.csv")

# Export cleaned dataframes to CSV files
customer_df.to_csv(customer_path, index=False)
print(f"✓ Saved: {customer_path}")

date_df.to_csv(date_path, index=False)
print(f"✓ Saved: {date_path}")

device_df.to_csv(device_path, index=False)
print(f"✓ Saved: {device_path}")

facts_df.to_csv(facts_path, index=False)
print(f"✓ Saved: {facts_path}")

product_df.to_csv(product_path, index=False)
print(f"✓ Saved: {product_path}")

print("\n✓ All cleaned dataframes have been exported ")
```
